## Función de onda variacional

Se propone una función de onda variacional con la siguiente forma

$\Psi_{\lambda} (r) = \Phi_{\mbox{fg}} F_\lambda = \Phi_{\mbox{fg}} e^{-\lambda r^{\frac32}} $

Siendo $\Phi_{\mbox{fg}}$ un término asociado a un gas de Fermi (determinante de Slater). $r$ corresponde a la distancia relativa entre dos cuerpos (cuando hay interacción).

Para el caso de 2 cuerpos unidos por un potencial lineal $V(r) = kr = k \left\vert r_1 - r_2 \right\vert$, se pueden calcular las siguientes cosas:

Calculamos la constante de normalización:

$\left\vert A \right\vert^2 = \left( (4 \pi) \int_0^\infty r^2 F_\lambda^2 dr \right)^{-1} = \frac{3 \lambda^2}{2 \pi}$

Recordando que para la parte radial

$F_\lambda^{'} = \frac{1}{r^2} \frac{\partial}{\partial r} \left( r^2 \frac{\partial}{\partial r} F_\lambda\right) = \frac{1}{r^2}\left(-\frac{15}{4} e^{-\lambda r^{3/2}} \lambda r^{3/2} + \frac94 e^{-\lambda r^{3/2}} \lambda^2 r^3 \right)$

Entonces, en general tenemos que para calcular el valor esperado de la energía total $\left\langle H \right\rangle_{\lambda} = \left\langle W \right\rangle_\lambda + \left\langle V \right\rangle_\lambda$, siendo $W$ un término cinético y $V$ un término potencial.

Usando unidades donde $\hbar = 1$ y siendo $\mu = \frac{m_1 m_2}{m_1 + m_2}$ la masa reducida.

$\left\langle W \right\rangle_\lambda = \left\vert A \right\vert^2 \left( 4 \pi \right) \left( -\frac{1}{2 \mu} \right) \int_0^\infty r^2 F_\lambda F_\lambda^{'} dr = \frac{9 \lambda^{4/3} \Gamma \left[\frac83 \right] }{8 (2^{2/3}) \mu}$

$\left\langle V \right\rangle_\lambda = \left\vert A \right\vert^2 \left( 4 \pi \right) \left( k \right) \int_0^\infty r^3 F_\lambda^2 dr = \frac{k \Gamma \left[ \frac83 \right]}{2^{2/3} \lambda^{2/3}}$

Con esto encontramos que para $k = 1, \ \mu = \frac12$, el valor del parámetro variacional óptimo es

$\lambda_0 = \frac{\sqrt{2}}{3}$

Siendo $\Psi_\lambda = \Phi_{fg} e^{-\lambda r^{\frac32}}$, tenemos que la condición del método MonteCarlo sería

$\frac{\left\vert \Psi_t \right\vert^2}{\left\vert \Psi_i \right\vert^2} = 1 = \frac{\Phi_{t} e^{-\lambda D_t}}{\Phi_{i} e^{-\lambda D_i}}$



### Radio al cuadrado

El valor esperado del radio al cuadrado de una partícula, con respecto al centro de masa del sistema del cual forma parte, lo podemos definir de la siguiente manera, para un sistema de mesones,

$\langle r_M^2 \rangle \equiv \langle \sum_i (\vec{r}_i - \vec{R})^2 \rangle$

siendo $\vec{R}$ el vector de centro de masa. En el caso de un meson con partículas 1 y 2 tenemos que

$\langle r_M^2 \rangle = ((\frac{m1}{m1 + m2})^2 + (\frac{m2}{m1+m2})^2) (\vec{r}_1 - \vec{r}_2)^2 = \frac{m_1^2 + m_2^2}{(m1 + m2)^2} \langle r^2 \rangle$

### Correlación. Función de distribución radial.

Las funciones de distribución radial son esencialmente histograma de distancia entre dos partículas.

a) Tomar una partícula $i$ de referencia, con posición $\vec{r}_i$

b) Dibujar una capa esférica de radio $r$ y grosor $\Delta r$ con centro en $\vec{r}_i$.

c) Una partícula $j$ en esta capa tiene una distancia $r_{ij} = \vert \vec{r}_i - \vec{r}_j \vert$, donde

$r - \Delta r \le r_{ij} < r$

Se determina el numero de partículas en la capa, llamando a este número $n_i (r,\Delta r)$.

d) Se divide por el volumen de la capa y se realiza un promedio sobre el número de partículas de referencia

$\frac{1}{N} \sum_i^N  \frac{n_i (r,\Delta r)}{4 \pi r^2 \Delta r}$

e) Se normaliza con respecto  a la densidad de partículas $\rho = N/V$ y se obtiene

$g(r) = \frac{V}{4 \pi r^2 \Delta r N^2} \sum_i^N n_i (r, \Delta r)$
